## Reto 1
Hay una caja con 10 bolas blancas, 12 bolas rojas y 8 bolas negras. Calcula la probabilidad de:
* Sacar una bola blanca.
* Sacar una bola blanca en la primera extracci√≥n y una bola negra en la segunda extracci√≥n sin reemplazar la primera.
* Sacar una bola blanca en la primera extracci√≥n y una bola negra en la segunda extracci√≥n reemplazando la primera.
* Sacar 3 bolas y que sean todas de diferentes colores.

```python
# Datos
blancas = 10
rojas = 12
negras = 8
total = blancas + rojas + negras

# F√ìRMULAS DEDUCIDAS EN LOS V√çDEOS ADJUNTOS Y PDF

print(f"La probabilidad de sacar una bola blanca es P(B) = {blancas / total}")

print(f"La probabilidad de sacar una bola blanca en la primera extracci√≥n y una bola negra en la segunda extracci√≥n sin reemplazar es P(B_1 ‚à© N_2) = {blancas/total * negras/(total-1)}")

print(f"La probabilidad de sacar una bola blanca en la primera extracci√≥n y una bola negra en la segunda extracci√≥n con reemplazo es P(B_1 ‚à© N_2) = {blancas/total * negras/total}")

print(f"La probabilidad de sacar 3 bolas de distintos colores sin remplazo es = {6 * blancas/total * rojas/total * negras/total}")
print(f"La probabilidad de sacar 3 bolas de distintos colores con remplazo es = {6 * (blancas/total * rojas/(total-1) * negras/(total-2))}")
```

## Reto 2

Quieres ir de picnic el s√°bado. Odias la lluvia y quieres planificar tu fin de semana. Para ayudarte a tomar una decisi√≥n, recopilas los siguientes datos sobre los d√≠as lluviosos:

* El 40% de todos los d√≠as est√°n nublados.
* Aproximadamente el 50% de los d√≠as nublados son lluviosos.
* Aproximadamente el 20% de los d√≠as no nublados son lluviosos.

¬øCu√°l es la probabilidad de que llueva el pr√≥ximo s√°bado?

```python
# Datos
PNsi = 0.4 # P(Nublado_si)
PLsiNsi = 0.5 # P(Llueve_si | Nublado_si)
PLsiNno = 0.2 # P(Llueve_si | Nublado_no)

# F√ìRMULAS DEDUCIDAS EN LOS V√çDEOS ADJUNTOS Y PDF

print(f"La probabilidad de que llueva el pr√≥ximo s√°bado es P(L) = {PNsi * PLsiNsi + (1-PNsi) * PLsiNno}")
```

## Reto 3

A y B son eventos de una probabilidad tal que \(P(A) = 0.3\), \(P(B) = 0.6\) y \(P(A ‚à© B) = 0.1\)

¬øCu√°les de las siguientes afirmaciones son verdaderas?
**SUGERENCIA: dibuja un gran diagrama de Venn a tu lado**
* \(P(¬¨{A}) = 0.7\)
* \(P(A ‚à™ B) = 0.6\)
* \(P(A ‚à© ¬¨ùêµ) = 0.2\)
* \(P(A ‚à™ ¬¨{A}) = 1\)  
* \(P(A ‚à© (B ‚à™ ¬¨{B})) = 0.4\)
* \(P(¬¨{A} ‚à© ¬¨{B}) = 0.3\)
* \(P(¬¨{A ‚à© B}) = 0.9\)

```python
# Datos
PA = 0.3
PB = 0.6
PAinterB = 0.1

# RESPUESTAS EXPLICADAS EN PDF CON DIAGRAMA DE VENN O F√ìRMULAS

print("Se calcula respuesta correcta con la f√≥rmula a partir de los datos y ")
print(f"se imprime {True} si es igual que la del enunciado o {False} si es distinta al enunciado. \n")

print(f"P(¬¨A) = {1-PA} ", 1-PA==0.7)
print(f"P(A ‚à™ B) = {round(PA+PB-PAinterB,2)} ", round(PA+PB-PAinterB,2) == 0.6)
print(f"P(A ‚à© ¬¨B) = {round(PA-PAinterB,2)} ", round(PA-PAinterB,2) == 0.2)
print(f"P(A ‚à™ ¬¨A) = {PA + (1-PA)} ", PA + (1-PA) == 1)
print(f"P(A ‚à© (B ‚à™ ¬¨B)) = {PA * 1} ", PA * 1 == 0.4)
print(f"P(¬¨A ‚à© ¬¨B) = {round(1 - (PA+PB-PAinterB),2)} ", round(1 - (PA+PB-PAinterB),2) == 0.3)
print(f"P(¬¨A‚à©B) = {1-PAinterB} ", 1-PAinterB == 0.9)
```

## Reto 4

* Si alguien viaja a China, la probabilidad de contraer Covid es del 1%
* Si alguien viaja a Italia, la probabilidad de contraer Covid es del 20%

Lanzo un dado con n√∫meros 1, 2, 3, 4, 5, 6:
* Si sale 1: Vuelo a Italia
* Si no sale 1: Vuelo a China

Durante mi viaje, contraje Covid. ¬øCu√°l es la probabilidad de que haya volado a China?

```python
# Datos
PI=1/6 # P(I)
PC=1-PI # P(C)
PCsi=0.01 # P(C|s√≠ covid)
PIsi=0.2 # P(I|s√≠ covid)

# F√ìRMULAS DEDUCIDA Y EXPLICADA EN PDF Y V√çDEO

print(f"La probabilidad de que haya volado a China sabiendo que contrajo covid es P(C|s√≠) = {PC * PCsi / (PC * PCsi + PI * PIsi)}")
```

## Simulado para dados de 6 caras

```python
import random
num_lanzamientos = 10000
n = 6 # n√∫mero de caras del dado

def lanzar_dado(n:int) -> int:
  """ Retorna el resultado de un dado de n caras.
  Entrada: n(int) = n√∫mero de caras del dado
  Salida: (int) resultado del dado entre 1 y n"""
  return random.randint(1, n)

# Inicializaci√≥n de posibles resultados de la suma de 2 dados de n caras.
suma_probabilidades = {resultado : 0 for resultado in range(2, 2*n+1)}
suma_frecuencias = {resultado : 0 for resultado in range(2, 2*n+1)}

# Para cada lanzamiento, se genera un n√∫mero aleatorio para cada dado, se suma el resultado y se incrementa en 1 en el diccionario esa suma.
for _ in range(num_lanzamientos):
  dado1 = lanzar_dado(n)
  dado2 = lanzar_dado(n)
  suma_dados = dado1 + dado2
  suma_frecuencias[suma_dados] += 1

# Calculo de la probabilidad dividiendo la frecuencia de cada resultado por el n√∫mero de lanzamientos total
for resultado in range(2, 2*n+1):
  suma_probabilidades[resultado] = suma_frecuencias[resultado] / num_lanzamientos

# Imprimir resultados
print("suma_frecuencias    ", suma_frecuencias)
print("suma_probabilidades ", suma_probabilidades)
```

```python
import matplotlib.pyplot as plt

plt.bar(suma_probabilidades.keys(), suma_probabilidades.values())
plt.show()
```

## Te√≥rico para dados de 6 caras

```python
def crear_lista(n):
  """ Crea una lista de naturales ascendente de 1 hasta el valor n que desciende hasta el 1 nuevamente.
  Esto son las combinaciones posibles de dos dados de ir obteniendo un n√∫mero de suma mayor, o sea la distribuci√≥n triangular."""
  ascendente = list(range(1, n + 1))  # Lista ascendente: 1 a n
  descendente = list(range(n - 1, 0, -1))  # Lista descendente: n-1 a 1
  return ascendente + descendente  # Combinar ambas listas

""" Se representan las sumas posibles range(2, 2*n+1)
 en frente a la probabilidad te√≥rica de dichas sumas
 Esa probabilidad se obtiene dividiendo las combinaciones de crear_lista con los casos posibles de n*n = n**2 resultados totales"""
plt.bar(list(range(2, 2*n+1)), [x / (n**2) for x in crear_lista(n)])
plt.xticks(range(2, 2*n + 1, 2))
plt.show()
```
